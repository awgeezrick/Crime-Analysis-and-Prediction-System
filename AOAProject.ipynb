```python
from tkinter import *
from PIL import Image, ImageTk
import pandas as pd
import numpy as np
import folium
from folium import plugins
from folium.plugins import MarkerCluster
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
# Set pandas's max row display
pd.set_option('display.max_row', 1000)
# Set pandas's max column width to 50
pd.set_option('display.max_columns', 50)
# Load in the Chicago crime dataset
df = pd.read_csv("C:/Users/Asif/Downloads/Crimes_-_2001_to_present.csv")
#print(df.head())


'''==========================================================================================='''
root = Tk()
root.geometry("995x500+120+120")
root.title("Crime Analysis and Prediction")
Tops = Frame(root, width=400, height=50, relief=SUNKEN)
Tops.pack(side=TOP, fill="y")
Label(Tops, text="Welcome to Crime Analysis and Prediction System\n", font="arial 31 bold", bg="#6666ff", pady=15,
      anchor='ne').grid(row=0, column=0)
Label(Tops, text="").grid(row=1, column=3)

'''===========================================DataSet==============================================='''
def crimedataset():
    def show():
        seriesObj=df.head(10)
        dicObj=seriesObj.values.tolist()
        #print(dicObj)
        listBox.insert(END, "Id| Case Number|Date                   |Type                |Loaction  | Arrest  | Beat |District")
        listBox.insert(END,"\n")
        for i in range(len(dicObj)):
            listBox.insert(END,dicObj[i][0])
            listBox.insert(END,"| ")
            listBox.insert(END,dicObj[i][1])
            listBox.insert(END,"|")
            listBox.insert(END,dicObj[i][2])
            listBox.insert(END,"|  ")
            listBox.insert(END,dicObj[i][5])
            listBox.insert(END,"|  ")
            listBox.insert(END,dicObj[i][7])
            listBox.insert(END,"|  ")
            listBox.insert(END,dicObj[i][8])
            listBox.insert(END,"|  ")
            listBox.insert(END,dicObj[i][10])
            listBox.insert(END,"|  ")
            listBox.insert(END,dicObj[i][11])

            listBox.insert(END,"\n")
           
    root11 = Toplevel()
    root11.geometry("995x500+120+120")
    root11.title("DataSet")

    listBox= Text(root11,width = 300, padx=50)
    listBox.grid(row = 1,column= 0, columnspan = 2)
    show()  
    #showPrediction = Button(root3, text = "Show Prediction",width = 15, command = show).grid(row = 4, column = 0)
    closeButton = Button(root11, text = "Close",width = 15, command = root11.destroy).grid(row = 4, column = 0)       
'''================================================================================================='''

'''============================Crime Rates====================================='''
def crimeRates():
    def show():
        crime_types=df['Primary Type'].value_counts()
        dic1=crime_types.to_dict()
        lst1=[]
        for i in dic1:
            lst2=list()
            l=dic1[i]
            lst2.append(i)
            lst2.append(l)
            #print(lst2)
            lst1.append(lst2)
        lst1.sort(key=lambda e: e[1], reverse=True)
        listBox.insert(END, "Id   |  Year           |  Total Crimes")

        listBox.insert(END,"\n")

        for i in range(len(lst1)):
            listBox.insert(END,(i+1))
            listBox.insert(END,"    |  ")
            listBox.insert(END,lst1[i][0])
            listBox.insert(END,"             |  ")
            listBox.insert(END,lst1[i][1])
            listBox.insert(END,"\n")
            
    root12 = Toplevel()
    root12.geometry("995x500+120+120")
    root12.title("Crime Rate")
    listBox= Text(root12,width = 200, padx=50)
    listBox.grid(row = 1,column= 0, columnspan = 2)
    show()  
    #showPrediction = Button(root3, text = "Show Prediction",width = 15, command = show).grid(row = 4, column = 0)
    closeButton = Button(root12, text = "Close",width = 15, command = root12.destroy).grid(row = 4, column = 1)



'''============================End Crime Rate Function=============='''


'''=============================Graph of Crimes========================='''
def graphCrime():
    
    ## Some formatting to make it look nicer
#     fig=plt.figure(figsize=(18, 16))
#     plt.title("Frequency of Crimes Per Crime Type")
#     plt.xlabel("Frequency of Crimes")
#     plt.ylabel("Type of Crime")
#     ax = crime_type_df.plot(kind='barh')
#     ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
    import tkinter as tk
    crime_type_df = df['Primary Type'].value_counts(ascending=True)
    root13=tk.Toplevel()
    figure1 = plt.Figure(figsize=(6, 5), dpi=100)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, root13)
    bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    crime_type_df.plot(kind='bar', legend=True, ax=ax1)
    ax1.set_title('Crime Vs.Number')

'''=============================End of Graph of Crimes===================='''

'''============================Crime Per Year   =========================='''

def crimeYear():
    def show():
        year=df['Year'].value_counts()
        dic1=year.to_dict()
        lst1=[]
        for i in dic1:
            lst2=list()
            l=dic1[i]
            lst2.append(i)
            lst2.append(l)
            #print(lst2)
            lst1.append(lst2)
        lst1.sort(key=lambda e: e[1], reverse=True)
        listBox.insert(END, "Id   |  Year    |  Total Crimes")

        listBox.insert(END,"\n")

        for i in range(len(lst1)):
            listBox.insert(END,(i+1))
            listBox.insert(END,"    |  ")
            listBox.insert(END,lst1[i][0])
            listBox.insert(END,"    |  ")
            listBox.insert(END,lst1[i][1])
            listBox.insert(END,"\n")
            
        listBox.insert(END, "\n")
        listBox.insert(END, "\n")
    root14= Toplevel()
    root14.geometry("995x500+120+120")
    root14.title("Crime Analysis and Prediction")
    label = Label(root14, text="Crime Per Year", font = ("Arial",30)).grid(row = 0, columnspan = 3)
    listBox= Text(root14,width = 100, padx=50)
    listBox.grid(row = 1,column= 0, columnspan = 2)
    show()  
    #showPrediction = Button(root3, text = "Show Prediction",width = 15, command = show).grid(row = 4, column = 0)
    closeButton = Button(root14, text = "Close",width = 15, command = root14.destroy).grid(row = 4, column = 1)

    
'''============================End Crime Per Year========================='''


def page1():
    root1=Toplevel()
    root1.geometry("995x500+120+120")
    root1.title("Crime Analysis and Prediction")
    Tops1 = Frame(root1, width=400, height=40, relief=SUNKEN)
    Tops1.pack(side=TOP, fill="y")
    Label(Tops1, text="Crime Analysis\n", font="arial 25 bold", bg="sky blue", pady=20, width=50 ).grid(row=0, column=0)
    Label(Tops1, text="").grid(row=1, column=3)
    b1 = Button(root1, text="View Dataset", font="arial 10 bold", width=25, fg="blue", bg="sky blue", border=6, command=crimedataset)
    #Label(Tops1, text="").grid(row=1, column=3)
    b1.pack()
    b2 = Button(root1, text="View Crime Rates", font="arial 10 bold", width=25, fg="blue", bg="sky blue", border=6, command=crimeRates)
    b2.pack()
    b4 = Button(root1, text="Graph of Crimes", font="arial 10 bold", width=25, fg="blue", bg="sky blue", border=6, command=graphCrime)
    b4.pack()
    b3 = Button(root1, text="Frequencies of Crimes", font="arial 10 bold", width=25, fg="blue", bg="sky blue", border=6, command=crimeYear)
    b3.pack()
    b5 = Button(root1, text="Exit", font="arial 10 bold", width=25, fg="blue", bg="sky blue", border=6,
                command=root1.destroy)
    b5.pack()



#Button for Crime Analysis
b1 = Button(root, text="Crime Analysis", font="arial 15 bold", width=15, fg="blue", bg="sky blue", border=6,
            command=page1)
Label(Tops, text="").grid(row=1, column=3)
b1.pack(pady=20)


def page2():
    root2 = Toplevel()
    root2.geometry("995x500+120+120")
    root2.title("Crime Analysis and Prediction")
    Tops2 = Frame(root2, width=400, height=40, relief=SUNKEN)
    Tops2.pack(side=TOP, fill="y")
    Label(Tops2, text="Criminal Record\n", font="arial 25 bold", bg="sky blue", pady=5, width=25).grid(row=0, column=0)
    Label(Tops2, text="").grid(row=1, column=3)
    b1 = Button(root2, text="Enter New Criminal Record ", font="arial 15 bold", width=25, fg="blue", bg="sky blue",
                border=6)
    Label(Tops2, text="").grid(row=1, column=3)
    b1.pack(pady=20)
    b2 = Button(root2, text="Search Data", font="arial 15 bold", width=25, fg="blue", bg="sky blue", border=6)
    b2.pack(pady=20)
    b3 = Button(root2, text="close", font="arial 15 bold", width=25, fg="blue", bg="sky blue", border=6,
                command=root2.destroy)
    b3.pack(pady=20)


b2 = Button(root, text="Criminal Data", font="arial 15 bold", width=15, fg="blue", bg="sky blue", border=6,
            command=page2)
b2.pack(pady=10)

def page3():
    from random import seed
    from random import randrange
    from csv import reader
    from math import sqrt
 
    # Load a CSV file
    def load_csv(filename):
        dataset = list()
        with open(filename, 'r') as file:
            csv_reader = reader(file)
            for row in csv_reader:
                if not row:
                    continue
                dataset.append(row)
        return dataset
 
    # Convert string column to float
    def str_column_to_float(dataset, column):
        for row in dataset:
            row[column] = float(row[column].strip())
 
    # Split a dataset into a train and test set
    def train_test_split(dataset, split):
        train = list()
        train_size = split * len(dataset)
        dataset_copy = list(dataset)
        while len(train) < train_size:
            index = randrange(len(dataset_copy))
            train.append(dataset_copy.pop(index))
        return train, dataset_copy
 
    # Calculate root mean squared error
    def rmse_metric(actual, predicted):
        sum_error = 0.0
        for i in range(len(actual)):
            prediction_error = predicted[i] - actual[i]
            sum_error += (prediction_error ** 2)
        mean_error = sum_error / float(len(actual))
        return sqrt(mean_error)
 
    # Evaluate an algorithm using a train/test split
    def evaluate_algorithm(dataset, algorithm, split, *args):
        train, test = train_test_split(dataset, split)
        test_set = list()
        for row in test:
            row_copy = list(row)
            row_copy[-1] = None
            test_set.append(row_copy)
        predicted = algorithm(train, test_set, *args)
        actual = [row[-1] for row in test]
        rmse = rmse_metric(actual, predicted)
        return rmse
 
    # Calculate the mean value of a list of numbers
    def mean(values):
        return sum(values) / float(len(values))
 
    # Calculate covariance between x and y
    def covariance(x, mean_x, y, mean_y):
        covar = 0.0
        for i in range(len(x)):
            covar += (x[i] - mean_x) * (y[i] - mean_y)
        return covar
 
    # Calculate the variance of a list of numbers
    def variance(values, mean):
        return sum([(x-mean)**2 for x in values])
 
    # Calculate coefficients
    def coefficients(dataset):
        x = [row[0] for row in dataset]
        y = [row[1] for row in dataset]
        x_mean, y_mean = mean(x), mean(y)
        b1 = covariance(x, x_mean, y, y_mean) / variance(x, x_mean)
        b0 = y_mean - b1 * x_mean
        return [b0, b1]
 
    # Simple linear regression algorithm
    def simple_linear_regression(train, test):
        predictions = list()
        b0, b1 = coefficients(train)
        for row in test:
            yhat = b0 + b1 * row[0]
            predictions.append(yhat)
        return predictions
 
    # Simple linear regression on insurance dataset
    seed(1)
    # load and prepare data
    filename = 'C:/Users/Asif/Documents/Prediction1.csv'
    dataset = load_csv(filename)
    for i in range(len(dataset[0])):
        str_column_to_float(dataset, i)
    # evaluate algorithm
    split = 0.6
    rmse = evaluate_algorithm(dataset, simple_linear_regression, split)
    print('RMSE: %.3f' % (rmse))
    def show():
        year=df['Year'].value_counts()
        dic1=year.to_dict()
        lst1=[]
        for i in dic1:
            lst2=list()
            l=dic1[i]
            lst2.append(i)
            lst2.append(l)
            #print(lst2)
            lst1.append(lst2)
        lst1.sort(key=lambda e: e[1], reverse=True)
        listBox.insert(END, "Id   |  Year    |  Total Crimes")

        listBox.insert(END,"\n")

        for i in range(len(lst1)):
            listBox.insert(END,(i+1))
            listBox.insert(END,"    |  ")
            listBox.insert(END,lst1[i][0])
            listBox.insert(END,"    |  ")
            listBox.insert(END,lst1[i][1])
            listBox.insert(END,"\n")
            
        listBox.insert(END, "\n")
        listBox.insert(END, "\n")
        listBox.insert(END, "Predicted Value of Total Crime is "+str(rmse)+".")
     
    root3 = Toplevel()
    root3.geometry("995x500+120+120")
    root3.title("Crime Analysis and Prediction")
    label = Label(root3, text="Prediction", font = ("Arial",30)).grid(row = 0, columnspan = 3)
    listBox= Text(root3,width = 100, padx=50)
    listBox.grid(row = 1,column= 0, columnspan = 2)
    show()  
    #showPrediction = Button(root3, text = "Show Prediction",width = 15, command = show).grid(row = 4, column = 0)
    closeButton = Button(root3, text = "Close",width = 15, command = root3.destroy).grid(row = 4, column = 1)



b3 = Button(root, text="Prediction", font="arial 15 bold", width=15, fg="blue", bg="sky blue", border=6, command=page3)
b3.pack(pady=10)

b4 = Button(root, text="close", font="arial 15 bold", width=15, fg="blue", bg="sky blue", border=6,
            command=root.destroy)
b4.pack(pady=10)


root.mainloop()
```
